```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset (assuming you have train.csv)
train_df = pd.read_csv("train.csv")
```

```python
# Step 1: Data Cleaning
# Step 1: Data Cleaning (safe version)
train_df['Age'] = train_df['Age'].fillna(train_df['Age'].median())
train_df['Embarked'] = train_df['Embarked'].fillna(train_df['Embarked'].mode()[0])
train_df.drop(columns=['Cabin'], inplace=True, errors='ignore')


# Create Age Groups
train_df['AgeGroup'] = pd.cut(train_df['Age'], bins=[0, 12, 18, 35, 60, 100],
                              labels=['Child', 'Teen', 'Young Adult', 'Adult', 'Senior'])

# Set Seaborn style
sns.set(style="whitegrid")
```

```python
# Plot 1: Survival count by Sex
plt.figure(figsize=(6, 4))
sns.countplot(x='Sex', hue='Survived', data=train_df)
plt.title('Survival Count by Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.tight_layout()
plt.show()
```

```python
# Plot 2: Survival count by Passenger Class
plt.figure(figsize=(6, 4))
sns.countplot(x='Pclass', hue='Survived', data=train_df)
plt.title('Survival Count by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.tight_layout()
plt.show()
```

```python
# Plot 3: Survival count by Age Group
plt.figure(figsize=(7, 4))
sns.countplot(x='AgeGroup', hue='Survived', data=train_df)
plt.title('Survival Count by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.tight_layout()
plt.show()
```

```python
# Plot 4: Survival count by Embarkation Point
plt.figure(figsize=(6, 4))
sns.countplot(x='Embarked', hue='Survived', data=train_df)
plt.title('Survival Count by Embarkation Point')
plt.xlabel('Port of Embarkation')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.tight_layout()
plt.show()
```
